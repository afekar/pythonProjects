```python
#test server
import time
import socket
import netifaces
from multiprocessing import Process
import random
import threading
import serial

ser = serial.Serial('/dev/ttyACM0',9600)
with open('file.txt','w') as f:
    f.write("")
def input_and_send():
    while 1:
        serialdata = ser.readline()
        serialdata = serialdata.decode("utf-8")
        print("distance: ",serialdata,"\n")
        with open('file.txt','a') as f:
            f.write(serialdata)

listener = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
listener.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)

iface = netifaces.gateways()['default'][netifaces.AF_INET][1]
IP = netifaces.ifaddresses(iface)[netifaces.AF_INET][0]['addr']
print(IP)
PORT = 12333
listener.bind((IP, PORT))
listener.listen(10)
while True:
    connection, address = listener.accept()
    proc2 = Process(target=input_and_send,daemon=True)
    proc2.start()
    while True:
        datain = connection.recv(1024).decode("utf-8")
        print(datain)
        print("")
        ser.write(datain.encode("utf-8"))
        if datain =='1':
            with open('file.txt','r') as f:
                connection.send(f.read().encode())
            with open('file.txt','w') as f:
                f.write("")
        if not datain:
            break
    proc2.terminate()
    connection.close()
```
